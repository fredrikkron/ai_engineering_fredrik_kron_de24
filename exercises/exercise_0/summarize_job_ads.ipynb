## Read job ads

```python
from google import genai

client = genai.Client()
job_ad_input = {"mime_type": "data/txt", "data": open("data/ads_1.txt", "rb").read()}

def ask_gemini(job_ad_input, model = "gemini-2.0-flash"):
    prompt = f"Summarize the job ad: {job_ad_input} in two sentences"
    response = client.models.generate_content(
        model=model,
        contents=prompt,
)
    return response

response = ask_gemini(job_ad_input)
print(response.text)
```
